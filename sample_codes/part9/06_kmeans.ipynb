K-means分群演算法

1.載入套件

```
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

2.隨機設定13個點

```
x=np.array([[1,1],[1,2],[2,1],[2,2],[3,3],[2,4],[4,2],[10,10],[9,8],[8,9],[7,7],[8,6],[6,8]])
```

3.建立K-means模型，可得分類結果

```
clf=KMeans(n_clusters=2)
clf.fit(x)

clf.labels_
```

4.繪出未分類的圖

```
plt.scatter(x[:,0],x[:,1],s=100)
plt.show()
```

5.繪出分類後的圖

```
plt.scatter(x[:,0],x[:,1],s=100,c=clf.labels_)
plt.show()
```

6.另一個範例，分成四類

```
x=np.random.rand(100,2)

clf=KMeans(n_clusters=2)
clf.fit(x)

plt.scatter(x[:,0],x[:,1],s=100,c=clf.labels_)
plt.show()
```
