決策樹-銀行貸款

1.載入套件

```
from sklearn import tree
```

2.設定X/Y軸資料，X軸為輸入資料，Y軸為是否借款

```
x=[[1,0,125],[1,1,100],[0,0,70],[1,1,120],[0,1,95],
   [0,1,60],[1,1,220],[0,0,85],[0,1,75],[0,0,90]]
y=[1,1,0,1,1,1,1,0,1,1]
```

3.建立決策樹模型

```
clf=tree.DecisionTreeClassifier()
clf=clf.fit(x,y)
```

4.進行預設

```
print(clf.predict([[1,0,10]]))
```

5.繪圖

```
tree.plot_tree(clf)
```
